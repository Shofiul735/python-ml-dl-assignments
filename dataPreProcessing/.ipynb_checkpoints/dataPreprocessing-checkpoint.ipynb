# Handling Missing values

```python
from sklearn.impute import SimpleImputer
```

```python
si_for_number = SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
dataset.iloc[:,0:1] = si_for_number.fit_transform(dataset.iloc[:,0:1])
```

```python
print(dataset.isnull().sum()) #count total null values
```

```python
dataset.iloc[2] # index 2 had age missing value, now the missing value is replaced by the mean value
```

```python
# most frequent value of the missing values' cloumns
gender_most_frequent_value = dataset['Gender'].mode().iloc[0]
weakness_most_frequent_value = dataset['weakness'].mode().iloc[0]

print(gender_most_frequent_value,weakness_most_frequent_value)
```

```python
si_for_string = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
```

```python
dataset.iloc[:,1:2] = si_for_string.fit_transform(dataset.iloc[:,1:2]) # age in cloumn 1
dataset.iloc[:,5:6] = si_for_string.fit_transform(dataset.iloc[:,5:6]) # weakness in cloumn 5
```

```python
dataset.iloc[22] # Gender information was missing in index 22
```

# Outliers

```python
print(dataset.nunique()) # we need to check outliers in the age column
```

```python
dataset.plot(kind='scatter', x='Age',y='Gender') # we have an outlier age = 600
```

```python
def drop_outlier_row(df,column_name, threshold=3):
    z_scores = (df[column_name] - df[column_name].mean()) / df[column_name].std()
    for index,value in z_scores.items():
        if (abs(value) > threshold):
            df.drop(index,inplace=True)
    return df
```

```python
dataset = drop_outlier_row(dataset,'Age')
```

```python
dataset.plot(kind='scatter', x='Age',y='Gender') # we had an outlier age = 600
```
